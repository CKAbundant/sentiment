# Bloomberg

- Use JJ's account to access Bloomberg website.
- Attempt to scrape info using Scrapling Fetcher, StealthyFetcher, and PlayWrightFetcher at "https://www.bloomberg.com/latest?utm_source=homepage&utm_medium=web&utm_campaign=latest"


````python
from scrapling import Fetcher, StealthyFetcher, PlayWrightFetcher
from dotenv import load_dotenv
import os

load_dotenv()
url = "https://www.bloomberg.com/latest?utm_source=homepage&utm_medium=web&utm_campaign=latest"
````

# Scrapling


## Fetcher

- Unable to post email and password to parse news info.


````python
fetcher = Fetcher(auto_match=False)

page = fetcher.post(url, email=os.getenv("USER"), password=os.getenv("PASSWORD"))
print(page.get_all_text())
````

## StealthyFetcher

- Unable to login to Bloomberg website despite passing 'extra_headers' argument.


````python
stealthyfetcher = StealthyFetcher()
page = await stealthyfetcher.async_fetch(
    url, extra_headers={"email": os.getenv("USER"), "password": os.getenv("PASSWORD")}
)
print(page.get_all_text())
````

## PlayWrightFetcher

- Able to parse info only by adding 'extra_headers' argument containing email and password.


````python
page = await PlayWrightFetcher().async_fetch(url)
print(page.get_all_text())
````

````python
page = await PlayWrightFetcher().async_fetch(
    url, extra_headers={"email": os.getenv("USER"), "password": os.getenv("PASSWORD")}
)
print(page.get_all_text())
````

# PlayWright


## Important Concepts

| Component                                                                             | Description                                                                                                                                                                                                                          |
| ------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------ |
| [Authentication](https://playwright.dev/python/docs/auth#signing-in-before-each-test) | &bull; Automate interaction with login form.                                                                                                                                                                                         |
| [Locators](https://playwright.dev/python/docs/locators)                               | &bull; Playwright's auto-waiting and retry-ability. <br>&bull; Represents a way to find element(s) on the page at any moment. <br>&bull; [List of roles in alphabetical order](https://www.w3.org/TR/wai-aria-1.2/#role_definitions) |
| [Navigation](https://playwright.dev/python/docs/navigations)                          | &bull; Navigate to webpage.                                                                                                                                                                                                          |
| [Other locators](https://playwright.dev/python/docs/other-locators)                   | &bull; Using non user visible locators (i.e. CSS or XPath) but could break when pages changes.                                                                                                                                       |
| [Pages](https://playwright.dev/python/docs/pages)                                     | &bull; Single tab or popup window within a broswer context. <br>&bull; Should be used to navigate to URLs and interact with page content.                                                                                            |

## Recommended built-in locators

| Locators                                                                                                                            | When to use                                                                                                                                                                                                                                                                          |
| ----------------------------------------------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------ |
| [page.get_by_role()](https://playwright.dev/python/docs/locators#locate-by-role)                                                    | &bull; Use whenever possible cos closest way to how users and assitive technology perceive the page.                                                                                                                                                                                 |
| [page.get_by_label()](https://playwright.dev/python/docs/locators#locate-by-label)                                                  | &bull; Locating form fields                                                                                                                                                                                                                                                          |
| [page.get_by_placeholder()](https://playwright.dev/python/docs/locators#locate-by-placeholder)                                      | &bull; Locating form elements that do not have labels but do have placeholder texts.                                                                                                                                                                                                 |
| [page.get_by_text()](https://playwright.dev/python/docs/locators#locate-by-text)                                                    | &bull; Locate non-interactive elements like `div`, `span`, `p`, etc. <br>&bull; For interactive elements like `button`, `a`, `input`, etc. use [role locators](https://playwright.dev/python/docs/locators#locate-by-role).                                                          |
| [page.get_by_alt_text()](https://playwright.dev/python/docs/locators#locate-by-alt-text)                                            | &bull; Use when your element supports alt text such as `img` and `area` elements.                                                                                                                                                                                                    |
| [page.get_by_title()](https://playwright.dev/python/docs/locators#locate-by-title)                                                  | &bull; Use when element has `title` attribute.                                                                                                                                                                                                                                       |
| [page.get_by_test_id()](https://playwright.dev/python/docs/locators#locate-by-test-id)                                              | &bull; Choose to use test id methodology. <br>&bull; Can't locate by [role](https://playwright.dev/python/docs/locators#locate-by-role) or [text](https://playwright.dev/python/docs/locators#locate-by-text).                                                                       |
| [page.locator(css=button") <br> page.locator("xpath=//button")](https://playwright.dev/python/docs/locators#locate-by-css-or-xpath) | &bull; CSS and XPath are not recommended as the DOM can often change leading to non resilient tests. <br>&bull; Instead, try to come up with a locator that is close to how the user perceives the page such as role locators or define an explicit testing contract using test ids. |

## Custom selector engines

- https://playwright.dev/python/docs/extensibility#custom-selector-engines
- Create a custom selector engines but must registered with selectors.register()

```
tag_selector = """
    // Must evaluate to a selector engine instance.
    {
      // Returns the first element matching given selector in the root's subtree.
      query(root, selector) {
        return root.querySelector(selector);
      },

      // Returns all elements matching given selector in the root's subtree.
      queryAll(root, selector) {
        return Array.from(root.querySelectorAll(selector));
      }
    }"""

# register the engine. selectors will be prefixed with "tag=".
playwright.selectors.register("tag", tag_selector)

# now we can use "tag=" selectors.
button = page.locator("tag=button")
button.click()

# we can combine it with built-in locators.
page.locator("tag=div").get_by_text("click me").click()

# we can use it in any methods supporting selectors.
button_count = page.locator("tag=button").count()
```

# Bloomberg

- Use JJ's account to access Bloomberg website.
- Attempt to scrape info using Scrapling Fetcher, StealthyFetcher, and PlayWrightFetcher at "https://www.bloomberg.com/latest?utm_source=homepage&utm_medium=web&utm_campaign=latest"


```python
from scrapling import Fetcher, StealthyFetcher, PlayWrightFetcher
from dotenv import load_dotenv
import os

load_dotenv()
url = "https://www.bloomberg.com/latest?utm_source=homepage&utm_medium=web&utm_campaign=latest"
```

# Fetcher

- Unable to post email and password to parse news info.


```python
fetcher = Fetcher(auto_match=False)

page = fetcher.post(url, email=os.getenv("USER"), password=os.getenv("PASSWORD"))
print(page.get_all_text())
```

# StealthyFetcher

- Unable to login to Bloomberg website despite passing 'extra_headers' argument.


```python
stealthyfetcher = StealthyFetcher()
page = await stealthyfetcher.async_fetch(
    url, extra_headers={"email": os.getenv("USER"), "password": os.getenv("PASSWORD")}
)
print(page.get_all_text())
```

# PlayWrightFetcher

- Able to parse info only by adding 'extra_headers' argument containing email and password.


```python
page = await PlayWrightFetcher().async_fetch(url)
print(page.get_all_text())
```

```python
page = await PlayWrightFetcher().async_fetch(
    url, extra_headers={"email": os.getenv("USER"), "password": os.getenv("PASSWORD")}
)
print(page.get_all_text())
```
